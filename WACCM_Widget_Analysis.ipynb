```python
# compute latitudinally weighted dataset from any 3D variable 
def LWAV(DS, time = False):
    try:
        DS = DS.mean(dim='lon')
    except:
        print('')
    if (time == False):
        try:
            DS = DS.mean(dim='time')
        except:
            print('')
    try:
        DS = gw*DS #multiply by gaussian weights
    except:
        DS = gw*DS.reshape(96,70) #multiply by gaussian weights
    try:
        DS = np.sum(DS,axis=1)/gw.sum() #weighted mean
    except:
        DS = np.sum(DS,axis=0)/gw.sum() #weighted mean 
    return DS #return a numpy array that has been modified above
```

```python
def SpecificLatWeight(DS,name,pos):
    if (name == None):
        try:
            DS = DS.mean(dim='lon')
        except:
            print('No longitude data for this dataset')
        try:
            DS = DS.mean(dim='time')
        except:
            print('No time series for this dataset')
        DS = DS[:70,47-pos:48+pos]*gw[47-pos:48+pos] #multiply by gaussian weights
        DS = np.sum(DS,axis=1)/gw[47-pos:48+pos].sum() #weighted mean
    else:
        try:
            DS = DS[name].mean(dim='lon')
        except:
            DS = DS[name]
            print('No longitude data for this dataset')
        try:
            DS = DS.mean(dim='time')
        except:
            print('No time series for this dataset')
        DS = DS[:70,47-pos:48+pos]*gw[47-pos:48+pos] #multiply by gaussian weights
        DS = np.sum(DS,axis=1)/gw[47-pos:48+pos].sum() #weighted mean
        
    return DS #return a numpy array that has been modified above
```

```python
def ComputePsiXr(v, t, lon='lon', lat='lat', pres='level', time='time', ref='rolling-91', p0=1e3):
    """Computes the residual stream function \Psi* (as a function of time).
    original code  by Martin Jucker from: https://github.com/mjucker/aostools

    INPUTS:
    v    - meridional wind, xr.DataArray
    t    - temperature, xr.DataArray
    lon    - name of longitude in t
    lat    - name of latitude in t
    pres    - name of pressure in t [hPa]
    time    - name of time field in t
    ref    - how to treat dTheta/dp: 
          - 'rolling-X' : centered rolling mean over X days
          - 'mean'	     : full time mean
    p0   - pressure basis to compute potential temperature [hPa]
    OUTPUTS:
    psi    - stream function, as a function of time
    psis    - residual stream function, as a function of time
    """
    from scipy.integrate import cumtrapz
    from numpy import cos,deg2rad
    # some constants
    kappa = 2./7
    a0    = 6371000
    g     = 9.81
    #
    ## compute psi
    v_bar,t_bar = ComputeVertEddyXr(v,t,pres,p0,lon,time,ref) # t_bar = bar(v'Th'/(dTh_bar/dp))

    # Eulerian streamfunction
    pdim = v_bar.get_axis_num(pres)
    psi = v_bar.reduce(cumtrapz,x=v_bar[pres],axis=pdim,initial=0) # [m.hPa/s]


    ## compute psi* = psi - bar(v'Th'/(dTh_bar/dp))
    psis = psi - t_bar
    coslat = np.cos(np.deg2rad(t[lat]))
    psi = 2*np.pi*a0/g*psi *coslat*100 #[kg/s]
    psis= 2*np.pi*a0/g*psis*coslat*100 #[kg/s]

    return psi,psis
```

```python
def day_night_terminator_contrast(DS, var, SubStellarLon, central_latitude = False):
    if (central_latitude == True):
        Sub_Stellar = DS[var][0,48:50,SubStellarLon].mean(dim = 'lat')
        if (SubStellarLon+36 >= 144):
            East_Terminator = DS[var][0,48:50,abs(144-(SubStellarLon+36))].mean(dim = 'lat')
        else:
            East_Terminator = DS[var][0,48:50,SubStellarLon+36].mean(dim = 'lat')
        if (SubStellarLon+72 >= 144):
            Anti_Stellar = DS[var][0,48:50,abs(144-(SubStellarLon+72))].mean(dim = 'lat')
        else:
            Anti_Stellar = DS[var][0,48:50,SubStellarLon+72].mean(dim = 'lat')
        if (SubStellarLon+108 >= 144):
            West_Terminator = DS[var][0,48:50,abs(144-(SubStellarLon+108))].mean(dim = 'lat')
        else:
            West_Terminator = DS[var][0,48:50,SubStellarLon+108].mean(dim = 'lat')
    else:
        Sub_Stellar = LWAV(DS[var][0,:,SubStellarLon])
        if (SubStellarLon+36 >= 144):
            East_Terminator = LWAV(DS[var][0,:,abs(144-(SubStellarLon+36))])
        else:
            East_Terminator = LWAV(DS[var][0,:,SubStellarLon+36])
        if (SubStellarLon+72 >= 144):
            Anti_Stellar = LWAV(DS[var][0,:,abs(144-(SubStellarLon+72))])
        else:
            Anti_Stellar = LWAV(DS[var][0,:,SubStellarLon+72])
        if (SubStellarLon+108 >= 144):
            West_Terminator = LWAV(DS[var][0,:,abs(144-(SubStellarLon+108))])
        else:
            West_Terminator = LWAV(DS[var][0,:,SubStellarLon+108])
    contrast = np.array([Sub_Stellar, East_Terminator, Anti_Stellar, West_Terminator])
    print('Values returned as:\nSub-Stellar point, East Terminator, Anti-Stellar Point, West Terminator')
    return contrast
```

```python
def compute_col(ds, field = 'O3', time = False, lon = True, g = 9.80616, m_var = False):
    
    P = ds1.lev.values
    h = 6.626*10**-34 # Planck's constant
    k = 1.381*10**-23 # boltzmann constant
    P_0 = 1e5 #surface pressure
    DU = (2.687*10**20) #1 Dobson Unit
    G = 6.67e-11 #Gravitational constant
    M_Earth = 5.972e24
    R_Earth = 6371e3
    
    #kboltz = 1.380044E-16   # boltzman's constant (erg/K)
    #R = 2.8704e6            # gas constant (erg/g/K)
    g = g             # grav acceleration (cm/sec2)
    #pa2dyn = 10.            # Pascals to dyn/cm2
    cm2todu = 1./2.6867e20  # convert colunm density / cm^2 to Dobson units
    m_avg = 28.96*1.661e-27
    
    #plev = ds['lev']
    
    #calcualte column integrals 
    variable = ds[field]#.values
    
    # variables to calcualate hybrid pressure on model interfaces
    ps = ds['PS'].values # surface pressure
    p0 = ds['P0'].values
    hyai = ds['hyai'].values
    hybi = ds['hybi'].values
    
    #col = np.ndarray(ps.shape, np.float32)
    dp = np.ndarray(variable.shape, np.float32)
    
    nt = ds['time'].size
    ny = ds['lat'].size
    nz = ds['lev'].size
    nx = ds['lon'].size
    
    #z = H*np.log(P/P[0])
    
    #g = g *(R_Earth**2)/(R_Earth+z)**2
    
    press = np.ndarray(nz, np.float32)
    for i in range(nt):
        for j in range(ny):
            for k in range(nx):
                press = hyai * p0 + hybi * ps[i,j,k]
                for iz in range(nz):
                    dp[i,iz,j,k] = (press[iz+1]-press[iz])
                    
    #calcualte column integrals 
    variable = ds[field]#.values
        
    variable = variable * dp / m_avg
    
    for i in range(nt):
        for j in range(96):
            for k in range(144):
                variable[i,:,j,k] = variable[i,:,j,k]/g
    
    if (field == 'O3'):
        var_col = np.sum(variable, axis = 1) * cm2todu
    else:
        var_col = np.sum(variable, axis = 1)
    if (lon == False):
        var_col  = var_col.mean(dim='lon')
    if (time == False):
        var_col  = var_col.mean(dim='time')
    print('Columns calculated')
    print('Returns number of molecules per metre squared in a column, or the number of Dobson units if O3 selected, or the mass in kg if m_var is not false but is a number (e.g. 18 for H2O)')
    if (m_var != False and field != 'O3'): var_col = var_col * m_var * 1.667e-27
    return var_col
```

```python
#%% Plot a single Earth projection - vectors

def zonal_mean_surface(DS,Var,level):
    if (len(ds1[Var].dims) > 3):
        DS = DS[Var].mean(dim = 'time')
        DS = DS.isel(lev = level)
    else:
        DS = DS[Var].mean(dim = 'time')
    return DS

def retrieve_data(var1, var2, ds, shape=(144, 96), lev = 69):
    """
    Returns ``(x, y, u, v, crs)`` of some vector data
    computed mathematically. The returned crs will be a rotated
    pole CRS, meaning that the vectors will be unevenly spaced in
    regular PlateCarree space.

    """
    crs = ccrs.RotatedPole(pole_longitude=0, pole_latitude=90)
    
    ds = ds
    
    lon = ds.lon.values[np.arange(0,144,4)]
    lat = ds.lat.values[np.arange(0,96,2)]
    
    U = zonal_mean_surface(ds,var1,lev).values[np.arange(0,96,2),:]
    U = U[:,np.arange(0,144,4)]
    V = zonal_mean_surface(ds,var2,lev).values[np.arange(0,96,2),:]
    V = V[:,np.arange(0,144,4)]

    return lon, lat, U, V, crs

def Plot_Earth_Vectors(lon = 0, lat = 0, plot_num = 1, fig_width = 5, fig_height = 5, lev = 69,
                       Projection = 'Orthographic', Borders = False, Land = False, Ocean = False, Grid = False, 
                       Rivers_Lakes = False):

    def Add_Earth_Features():
        zorder1 = 0; zorder2 = 0; Design = ''
        if (Ocean == True):
            zorder1 = 1; Design = '_Ocean'
        if (Land == True):
            zorder2 = 1; Design = '_Land'
        ax.add_feature(cartopy.feature.OCEAN, zorder=zorder1, facecolor = 'royalblue')
        ax.add_feature(cartopy.feature.LAND, zorder=zorder2, edgecolor='black', facecolor = 'forestgreen')

        if (Rivers_Lakes == True):
            import cartopy.feature as cfeature
            rivers_50m = cfeature.NaturalEarthFeature('physical', 'rivers_lake_centerlines', '50m')
            ax.add_feature(rivers_50m, facecolor='None', edgecolor='skyblue')
            ax.add_feature(cartopy.feature.RIVERS, zorder=zorder2, edgecolor = 'skyblue')#, facecolor = 'royalblue')
            ax.add_feature(cartopy.feature.LAKES, zorder=zorder2, edgecolor = 'skyblue', facecolor = 'skyblue')
        Border = ''
        if (Borders == True):
            Border = 'Borders'    
            ax.add_feature(cartopy.feature.BORDERS, zorder=zorder2, edgecolor = 'k')

        ax.set_global()
        if (Grid == True):
            ax.gridlines()
            
    Figure = plt.figure(figsize=(fig_width, fig_height))
    #change the longitude and latiude below of where you would like to centre the image
    var1 = 'U'; var2 = 'V'
    lon_centre = lon; lat_centre = lat
    ax = plt.subplot2grid((1,plot_num), (0,0), projection=ccrs.Orthographic(central_longitude=lon_centre, central_latitude=lat_centre))
    if (Projection == 'Robinson'): ax = plt.subplot2grid((1,plot_num), (0,0), projection=ccrs.Robinson(central_longitude=lon_centre))             
    if (Projection == 'Mollweide'): ax = plt.subplot2grid((1,plot_num), (0,0), projection=ccrs.Mollweide(central_longitude=lon_centre))
    if (Projection == 'Mercator'): ax = plt.subplot2grid((1,plot_num), (0,0), projection=ccrs.Mercator(central_longitude=lon_centre))
    if (Projection == 'EckertIV'): ax = plt.subplot2grid((1,plot_num), (0,0), projection=ccrs.EckertIV(central_longitude=lon_centre))          
    Add_Earth_Features()
    x, y, u, v, vector_crs = retrieve_data(var1 = var1, var2 = var2, ds = ds1, lev = lev)
    ax.quiver(x, y, u, v, transform=vector_crs, color = 'k', minshaft = 2, width =0.004, headwidth = 9)
    plt.title(ds1_name+ '\nmean winds\nlevel = ' +str(ds1.lev.values[lev])+' hPa', fontsize = 20)
    
    if (plot_num >= 2):
        ax = plt.subplot2grid((1,plot_num), (0,1), projection=ccrs.Orthographic(central_longitude=lon_centre, central_latitude=lat_centre))
        if (Projection == 'Robinson'): ax = plt.subplot2grid((1,plot_num), (0,1), projection=ccrs.Robinson(central_longitude=lon_centre))             
        if (Projection == 'Mollweide'): ax = plt.subplot2grid((1,plot_num), (0,1), projection=ccrs.Mollweide(central_longitude=lon_centre))
        if (Projection == 'Mercator'): ax = plt.subplot2grid((1,plot_num), (0,1), projection=ccrs.Mercator(central_longitude=lon_centre))
        if (Projection == 'EckertIV'): ax = plt.subplot2grid((1,plot_num), (0,1), projection=ccrs.EckertIV(central_longitude=lon_centre))          
        Add_Earth_Features()
        x, y, u, v, vector_crs = retrieve_data(var1 = var1, var2 = var2, ds = ds2, lev = lev)
        ax.quiver(x, y, u, v, transform=vector_crs, color = 'k', minshaft = 2, width =0.004, headwidth = 9)
        plt.title(ds2_name+ '\nmean winds\nlevel = '+str(ds1.lev.values[lev])+' hPa', fontsize = 20)

    if (plot_num >= 3):
        ax = plt.subplot2grid((1,plot_num), (0,2), projection=ccrs.Orthographic(central_longitude=lon_centre, central_latitude=lat_centre))
        if (Projection == 'Robinson'): ax = plt.subplot2grid((1,plot_num), (0,2), projection=ccrs.Robinson(central_longitude=lon_centre))             
        if (Projection == 'Mollweide'): ax = plt.subplot2grid((1,plot_num), (0,2), projection=ccrs.Mollweide(central_longitude=lon_centre))
        if (Projection == 'Mercator'): ax = plt.subplot2grid((1,plot_num), (0,2), projection=ccrs.Mercator(central_longitude=lon_centre))
        if (Projection == 'EckertIV'): ax = plt.subplot2grid((1,plot_num), (0,2), projection=ccrs.EckertIV(central_longitude=lon_centre))          
        Add_Earth_Features()
        x, y, u, v, vector_crs = retrieve_data(var1 = var1, var2 = var2, ds = ds3, lev = lev)
        ax.quiver(x, y, u, v, transform=vector_crs, color = 'k', minshaft = 2, width =0.004, headwidth = 9)
        plt.title(ds3_name+ '\nmean winds\nlevel = '+str(ds1.lev.values[lev])+' hPa', fontsize = 20)

    if (plot_num >= 4):
        ax = plt.subplot2grid((1,plot_num), (0,3), projection=ccrs.Orthographic(central_longitude=lon_centre, central_latitude=lat_centre))
        Add_Earth_Features()
        x, y, u, v, vector_crs = retrieve_data(var1 = var1, var2 = var2, ds = ds4, lev = lev)
        ax.quiver(x, y, u, v, transform=vector_crs, color = 'k', minshaft = 2, width =0.004, headwidth = 9)
        plt.title(ds4_name+ '\nmean winds\nlevel = '+str(ds1.lev.values[lev])+' hPa', fontsize = 20)

    if (plot_num >= 5):
        ax = plt.subplot2grid((1,plot_num), (0,4), projection=ccrs.Orthographic(central_longitude=lon_centre, central_latitude=lat_centre))
        Add_Earth_Features()
        x, y, u, v, vector_crs = retrieve_data(var1 = var1, var2 = var2, ds = ds5, lev = lev)
        ax.quiver(x, y, u, v, transform=vector_crs, color = 'k', minshaft = 2, width =0.004, headwidth = 9)
        plt.title(ds5_name+ '\nmean winds\nlevel = '+str(ds1.lev.values[lev])+' hPa', fontsize = 20)
```
